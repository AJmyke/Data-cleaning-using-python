# 6.3 Salary Distribution

```python
print(df.columns)
```

```python
df['salary_estimate'].head(10)
```

```python
import numpy as np
import pandas as pd

# Clean symbols and text
df['salary_estimate'] = (
    df['salary_estimate']
    .astype(str)
    .str.replace(r'[\$,Kk¬£]', '', regex=True)
    .str.replace('per year', '', case=False)
    .str.strip()
)

# Only split if at least one dash exists
if df['salary_estimate'].str.contains('-').any():
    split_salary = df['salary_estimate'].str.split('-', n=1, expand=True)
else:
    # Fallback: treat the whole value as both min and max
    split_salary = pd.DataFrame({
        0: df['salary_estimate'],
        1: df['salary_estimate']
    })

# Assign columns safely
df['Minimum_Salary'] = pd.to_numeric(split_salary[0], errors='coerce')
df['Maximum_Salary'] = pd.to_numeric(split_salary[1], errors='coerce')

# Fill any missing maximums with the minimum
df['Maximum_Salary'] = df['Maximum_Salary'].fillna(df['Minimum_Salary'])

# Compute average
df['avg_salary'] = (df['Minimum_Salary'] + df['Maximum_Salary']) / 2
```

```python
df[['salary_estimate', 'Minimum_Salary', 'Maximum_Salary', 'avg_salary']].head(10)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='avg_salary', bins=25, kde=True, color='steelblue')
plt.title('Distribution of Average Data Science Salaries', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Average Salary ($)', fontsize=12)
plt.ylabel('Number of Job Postings', fontsize=11)
sns.despine()
plt.tight_layout()
plt.show()
```

```python
median_salary = df['avg_salary'].median()
q1 = df['avg_salary'].quantile(0.25)
q3 = df['avg_salary'].quantile(0.75)

print(f"Median: ${median_salary:,.0f}")
print(f"25th‚Äì75th percentile range: ${q1:,.0f} ‚Äì ${q3:,.0f}")
```

# 6.4 Average Company Rating by Sector

```python
plt.figure(figsize=(10,6))

# Calculate mean rating per sector
sector_rating = df.groupby('sector')['rating'].mean().sort_values(ascending=False).head(10)

# Plot (with explicit hue)
sns.barplot(
    x=sector_rating.values,
    y=sector_rating.index,
    hue=sector_rating.index,      
    palette='viridis',
    legend=False                  
)

# Styling
plt.title('Top 10 Sectors by Average Company Rating', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Average Rating', fontsize=12)
plt.ylabel('Sector', fontsize=12)
sns.despine(left=True, bottom=True)
plt.tight_layout()
plt.show()
```

# üí° Insights on Company Rating by Sector

The Media and Information Technology sectors stand out with the highest average company ratings, suggesting stronger employee satisfaction and healthier workplace cultures.

Finance and Government sectors also maintain relatively high ratings, possibly reflecting better growth opportunities and competitive benefits.

On the other hand, Insurance and Manufacturing sectors show lower average ratings, which may stem from operational stress, lower compensation levels, or limited flexibility.

Overall, the results indicate that data-driven and innovation-oriented sectors tend to receive more favorable ratings, aligning with trends in modern workforce expectations.
