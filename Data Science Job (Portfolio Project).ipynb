# Dropping irrelevant columns

```python
# Dropping both index column and Competitors column, because they seem irrelevant and redundant.

df.drop(['index','Competitors'], axis = 1, inplace = True)
df.head()
```

# Removing irregular strings and characters

```python
df['Company Name'] = df['Company Name'].str.strip('\n3.1')
df['Company Name'] = df['Company Name'].str.strip('\n4')
df['Company Name'] = df['Company Name'].str.strip('\n2')
df['Company Name'] = df['Company Name'].str.strip('\n5.0')
df['Company Name'] = df['Company Name'].str.strip('\n3.6')
df['Company Name'] = df['Company Name'].str.strip('\n2.9')
df['Company Name'] = df['Company Name'].str.strip('\n3.8')
df['Company Name'] = df['Company Name'].str.strip('\n4.2')
df['Company Name'] = df['Company Name'].str.strip('\n3.5')
df['Company Name'] = df['Company Name'].str.strip('\n2.7')
df['Company Name'] = df['Company Name'].str.strip('\n4.7')
df['Company Name'] = df['Company Name'].str.strip('\n3.7')
df.head()
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.strip('(Glassdoor est.)')
df.head()
df['Salary Estimate'] = df['Salary Estimate'].str.strip('(Employer est.)')
df.head()
```

```python
df['Rating'] = df['Rating'].replace(-1.0,0)
df.tail(10)
```

```python
df.drop(df[df['Founded'] <= -1].index, inplace = True)
df.tail(10)
```

```python
df['Size'] = df['Size'].replace(-1,'Unknown')
df.head()
```

```python
df['Type of ownership'] = df['Type of ownership'].replace(-1,'Unknown')
df.head()
```

```python
df['Revenue'] = df['Revenue'].replace(-1,'Unknown / Non-Applicable')
df.head()
```

# Creating a new columns called Minimum_Salary and Maximum_Salary

```python
# Minimum_Salary

df['Minimum_Salary'] = df['Salary Estimate'].apply(lambda x: x[1:5])
```

```python
# Maximum_Salary

df['Maximum_Salary'] = df['Salary Estimate'].apply(lambda x: x[-4:])
```

```python
df[['Minimum_Salary','Maximum_Salary']].sample(10)
```

# Creating a new column called Job_State

```python
#Job_State

def check_location(val):
    if 'Remote' in val:
        return 'Remote'
    elif 'United States' in val:
        return 'US'
    else:
         return val[-3:]

df['Job_State'] = df['Location'].apply(lambda x: check_location(x))
```

```python
df['Job_State'].sample(10)
```
